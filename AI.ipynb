```python


import pandas as pd
from thefuzz import process

# 1Ô∏è‚É£ Load the dataset
df = pd.read_csv("fertilizer_price.csv")  # or use pd.read_excel("fertilizer_price.xlsx")

# Ensure the Price column is numeric
df['Price (??/50kg)'] = pd.to_numeric(df['Price (??/50kg)'], errors='coerce')

# Drop rows with missing price or approval info
df = df.dropna(subset=['Price (??/50kg)', 'Approved (Yes/No)'])

# Clean the 'Approved (Yes/No)' column: remove spaces and lowercase
df['Approved (Yes/No)'] = df['Approved (Yes/No)'].str.strip().str.lower()

# 2Ô∏è‚É£ Function to recommend fertilizer
def recommend_fertilizer(input_name, threshold=90):
    # Check exact match first
    exact_match = df[df['Fertilizer Name'].str.lower() == input_name.lower()]
    if not exact_match.empty:
        matched = exact_match.iloc[0]
    else:
        # Fuzzy match if exact not found
        result = process.extractOne(input_name, df['Fertilizer Name'])
        if result is None or result[1] < threshold:
            return "‚ùå Not approved by government / No matching fertilizer found."
        closest_match = result[0]
        score = result[1]
        matched = df[df['Fertilizer Name'] == closest_match].iloc[0]
        # Suggest the closest name to user
        print(f"üí° Did you mean '{closest_match}'? (Similarity: {score}%)")

    # Determine approval status
    approval_status = "‚úÖ Approved" if matched['Approved (Yes/No)'] == 'yes' else "‚ùå Not approved"
    price = matched['Price (??/50kg)']

    return f"{approval_status}\nPrice: ‚Çπ{price}"

# 3Ô∏è‚É£ Example usage
while True:
    user_input = input("Enter fertilizer name (or 'exit' to quit): ")
    if user_input.lower() == 'exit':
        break
    print(recommend_fertilizer(user_input))


```
